Criar o modelo

```python
df['Glucose_BMI'] = df['Glucose'] * df['BMI']
df['Glucose_Age'] = df['Glucose'] * df['Age']
df['BMI_Age'] = df['BMI'] * df['Age']
```

```python
X=df.drop('Outcome', axis=1)
y= df['Outcome']
```

```python
#important_features = ["Glucose", "BMI", "Age", "Pregnancies"]
```

```python
#X_important = X[important_features]
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=.2)
```

SMOTE

```python
#ros = RandomOverSampler(random_state=42)
#X_ros,y_ros = ros.fit_resample(X,y)

smote = SMOTE(random_state=42)
X_ros,y_ros = smote.fit_resample(X,y)
```

```python
sns.countplot(data=df,x=y_ros)
```

```python
X_trainOver, X_testOver, y_trainOver, y_testOver = train_test_split(X_ros, y_ros, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_trainOver = scaler.fit_transform(X_trainOver)
X_testOver = scaler.transform(X_testOver)
```

Modelo Decision Tree Classifier

```python
modeloDecisionTreeClas = DecisionTreeClassifier(max_depth=4,min_samples_leaf=40,min_samples_split=50,random_state=42)
```

```python
modeloDecisionTreeClas.fit(X_trainOver, y_trainOver)
```

```python
modeloDecisionTreeClasOver = modeloDecisionTreeClas.predict(X_testOver)
modeloDecisionTreeClasOver
```

```python
print("Importância das Features:")
print(modeloDecisionTreeClas.feature_importances_)
```

```python
accuracyDecOver = accuracy_score(y_testOver, modeloDecisionTreeClasOver)
recallDecOver = recall_score(y_testOver, modeloDecisionTreeClasOver)
fDecOver = f1_score(y_testOver, modeloDecisionTreeClasOver)
```

```python
print("Acuracia:", accuracyDecOver)
print("Recall:", recallDecOver)
print("F1 Score:", fDecOver)
```

```python
modeloDecisionTreeClasOver = confusion_matrix(y_testOver, modeloDecisionTreeClasOver)
modeloDecisionTreeClasOver
```

```python
train_sizes, train_scores, val_scores = learning_curve(
    modeloDecisionTreeClas, X_trainOver, y_trainOver, cv=5, scoring='accuracy', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', color='r', label='Acurácia de Treinamento')
plt.plot(train_sizes, val_mean, 'o-', color='g', label='Acurácia de Validação')

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='g')

plt.title('Curvas de Aprendizado do Modelo de Árvore de Decisão Decision Tree Classifier')
plt.xlabel('Tamanho do Conjunto de Treinamento')
plt.ylabel('Acurácia')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
y_pred_proba = modeloDecisionTreeClas.predict_proba(X_testOver)[:, 1]

# Calcule a curva ROC
fpr, tpr, thresholds = roc_curve(y_testOver, y_pred_proba)

# Calcule a AUC
roc_auc = auc(fpr, tpr)

# Plote a curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Falso Positivo')
plt.ylabel('Verdadeiro Positivo')
plt.title('Receiver Operating Characteristic (Árvore de Decisão)')
plt.legend(loc="lower right")
plt.show()
```

```python
scores = cross_val_score(modeloDecisionTreeClas, X_ros, y_ros, cv=5, scoring='accuracy')

print(f"Acurácia média com validação cruzada: {scores.mean():.2f}")
print(f"Desvio padrão da acurácia com validação cruzada: {scores.std():.2f}")
```
