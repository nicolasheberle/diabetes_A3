Criar o modelo

```python
X=df.drop('Outcome', axis=1)
y= df['Outcome']
```

```python
#important_features = ["Glucose", "BMI", "Age", "Pregnancies"]
```

```python
#X_important = X[important_features]
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=.2)
```

```python
#ros = RandomOverSampler(random_state=42)
#X_ros,y_ros = ros.fit_resample(X,y)

smote = SMOTE(random_state=42)
X_ros,y_ros = smote.fit_resample(X,y)
```

```python
sns.countplot(data=df,x=y_ros)
```

```python
X_trainOver, X_testOver, y_trainOver, y_testOver = train_test_split(X_ros, y_ros, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_trainOver = scaler.fit_transform(X_trainOver)
X_testOver = scaler.transform(X_testOver)
```

Random Forest com SMOTE

```python
modeloRandomForestOver = RandomForestClassifier(max_depth=2,min_samples_leaf=40, min_samples_split=30,random_state=42)
```

```python
modeloRandomForestOver.fit(X_trainOver,y_trainOver)
```

```python
parametros2 = {'bootstrap':[True, False],
              'criterion': ['gini','entropy','log_loss'],
              'max_features':['sqrt','log2'],
              'max_samples':[2,6,8,12]
}
```

```python
modeloRandom2 = GridSearchCV(RandomForestClassifier(),
                                parametros2,
                                scoring= 'accuracy',
                                n_jobs = -1)
```

```python
modeloRandom2.fit(X_trainOver, y_trainOver)
```

```python
melhorModelo = modeloRandom2.best_estimator_
```

```python
print(melhorModelo)
```

```python
modeloRandomOver = modeloRandomForestOver.predict(X_testOver)
modeloRandomOver
```

```python
train_sizes, train_scores, val_scores = learning_curve(
    modeloRandomForestOver, X_trainOver, y_trainOver, cv=5, scoring='accuracy', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', color='r', label='Acurácia de Treinamento')
plt.plot(train_sizes, val_mean, 'o-', color='g', label='Acurácia de Validação')

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='g')

plt.title('Curvas de Aprendizado do Modelo de Árvore de Decisão Random Forest')
plt.xlabel('Tamanho do Conjunto de Treinamento')
plt.ylabel('Acurácia')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
accuracyRandomUnd = accuracy_score(y_testOver, modeloRandomOver)
recallRandomUnd = recall_score(y_testOver, modeloRandomOver)
fRandomUnd = f1_score(y_testOver, modeloRandomOver)
```

```python
print("Acuracia:", accuracyRandomUnd)
print("Recall:", recallRandomUnd)
print("F1 Score:", fRandomUnd)
```

```python
class_reportRandomOver = classification_report(y_testOver, modeloRandomOver)
print("\nClassification Report:")
print(class_reportRandomOver)
```

```python
matrizRandomOver = confusion_matrix(y_testOver, modeloRandomOver)
matrizRandomOver
```

```python
y_pred_proba = modeloRandomForestOver.predict_proba(X_testOver)[:, 1]

# Calcule a curva ROC
fpr, tpr, thresholds = roc_curve(y_testOver, y_pred_proba)

# Calcule a AUC
roc_auc = auc(fpr, tpr)

# Plote a curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (Árvore de Decisão)')
plt.legend(loc="lower right")
plt.show()
```

```python
scores = cross_val_score(modeloRandomForestOver, X_ros, y_ros, cv=5, scoring='accuracy')

print(f"Acurácia média com validação cruzada: {scores.mean():.2f}")
print(f"Desvio padrão da acurácia com validação cruzada: {scores.std():.2f}")
```
