Criar o modelo

```python
df['Glucose_BMI'] = df['Glucose'] * df['BMI']
df['Glucose_Age'] = df['Glucose'] * df['Age']
df['BMI_Age'] = df['BMI'] * df['Age']
```

```python
X=df.drop('Outcome', axis=1)
y= df['Outcome']
```

```python
#important_features = ["Glucose", "BMI", "Age", "Pregnancies"]
```

```python
#X_important = X[important_features]
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=.2)
```

SMOTE

```python
#ros = RandomOverSampler(random_state=42)
#X_ros,y_ros = ros.fit_resample(X,y)

smote = SMOTE(random_state=42)
X_ros,y_ros = smote.fit_resample(X,y)
```

```python
sns.countplot(data=df,x=y_ros)
```

```python
X_trainOver, X_testOver, y_trainOver, y_testOver = train_test_split(X_ros, y_ros, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_trainOver = scaler.fit_transform(X_trainOver)
X_testOver = scaler.transform(X_testOver)
```

XGBoost

```python
dtreino = xgb.DMatrix(X_trainOver, label = y_trainOver)
dteste = xgb.DMatrix(X_testOver, label=y_testOver)
```

```python
parametro = {
    'max_depth': 1,       # Profundidade máxima da árvore
    'eta': 0.1,           # Taxa de aprendizado
    'objective': 'binary:logistic',  # Objetivo de classificação binária
    'eval_metric': 'logloss' ,        # Métrica de avaliação
    'alpha': 0.1,
    'lambda': 0.1
}
```

```python
num_round = 100
```

```python
bst = xgb.train(parametro, dtreino, num_round)
```

```python
preds = bst.predict(dteste)
predictions = [1 if pred > 0.5 else 0 for pred in preds]
```

```python
cm = confusion_matrix(y_testOver, predictions)
print(f"Matriz de confusão:\n{cm}")
```

```python
modelXGBoost = xgb.XGBClassifier(
    max_depth=5,
    learning_rate=0.002,
    n_estimators=50,
    objective='binary:logistic',
    eval_metric='logloss',
    reg_alpha=0.1,
    reg_lambda=0.1,
)
```

```python
modelXGBoost.fit(X_trainOver, y_trainOver)
```

```python
y_pred5 = modelXGBoost.predict(X_testOver)
```

```python
modelXGBmatriz = confusion_matrix(y_testOver, y_pred5)
modelXGBmatriz
```

```python
train_sizes, train_scores, val_scores = learning_curve(
    modelXGBoost, X_trainOver, y_trainOver, cv=5, scoring='accuracy', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', color='r', label='Acurácia de Treinamento')
plt.plot(train_sizes, val_mean, 'o-', color='g', label='Acurácia de Validação')

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='g')

plt.title('Curvas de Aprendizado do Modelo de Árvore de Decisão XGBoost')
plt.xlabel('Tamanho do Conjunto de Treinamento')
plt.ylabel('Acurácia')
plt.legend(loc='best')
plt.grid()
plt.show()
```
